```python
import numpy as np
import scipy.stats as stats
from scipy.stats import beta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a2, b2 = 0.7, 0.7
```

```python
x = np.linspace(beta.ppf(0.01, a2, b2), beta.ppf(0.99, a2, b2), 1000)
plt.plot(x, beta.pdf(x, a2, b2), 'r-', lw=2, alpha=0.6, label='beta pdf')
```

```python
a,b = 3,3
```

```python
x = np.linspace(beta.ppf(0.001, a, b), beta.ppf(0.999, a, b), 1000)
plt.plot(x, beta.pdf(x, a, b), 'r-', lw=2, alpha=0.6, label='beta pdf')
```

```python
r = beta.rvs(a, b, size=100000)
```

```python
sns.distplot(r)
```

```python
sns.distplot(beta.ppf(beta.cdf(r, a,b), a2,b2))
```

```python
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
```

```python
x = np.linspace(beta.ppf(0.0001, a, b), beta.ppf(0.9999, a, b), 10000)
y = beta.pdf(x, a, b)

points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
```

```python
eps = 0.05
fig, ax = plt.subplots()

# Create a continuous norm to map from data points to colors
norm = plt.Normalize(y.min(), y.max())
lc = LineCollection(segments, cmap='viridis', norm=norm)
# Set the values used for colormapping
lc.set_array(y)
lc.set_linewidth(2)
line = ax.add_collection(lc)
cbar = fig.colorbar(line, ax=ax)
cbar.set_ticks([])


ax.set_xlim(0.-eps, 1.+eps)
ax.set_ylim(0., y.max()+eps)
plt.show()
```
